```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import geopandas as gpd
```

# Подготовка визуализаций

```python
df = pd.read_excel('datasets/disorders.xlsx')
df.set_index(df.columns[0], inplace = True)
#df['EVENT_DATE'] = pd.to_datetime(df['EVENT_DATE'])
df.head()
```

Для географических визуализаций будем использовать библиотеку GeoPandas

```python
shape = gpd.read_file("mapshape/ne_110m_admin_0_countries.shp") #файл с картой
```

```python
df_sum = df[["UNORGANIZED", "ISO"]].groupby("ISO").sum()
df_sum["DISORDERS"] = df[["UNORGANIZED", "ISO"]].groupby("ISO").count()
df_sum["RATIO"] = df_sum["UNORGANIZED"] / df_sum["DISORDERS"]
df_sum.reset_index(inplace = True)
df_sum = pd.DataFrame(df_sum)
df_sum["ISO"].astype("int64")
df_sum
```

Теперь у нас есть файл, в котором наблюдения по странам просуммированы, и создали новый признак - соотношение "неполитических" событий ко всем.
Соединим файл наблюдений с файлом карты.

```python
df_geo = shape[["ISO_N3", "geometry"]]
df_geo["ISO"] = pd.to_numeric(df_geo["ISO_N3"])
df_geo.drop(labels = "ISO_N3", axis = 1, inplace = True)
df_geo = df_geo.merge(df_sum, on="ISO")
df_geo
```

Нарисуем пробную карту

```python
df_geo.plot(column = "RATIO", legend = True)
```

Даже на этой маленькой карте видно, что на нашей карте, вообще то, НЕТ ФРАНЦИИ И НОРВЕГИИ. А чего еще нет?

```python
df_geo = shape[["ISO_N3", "geometry", "NAME", "ADMIN"]]
df_geo["ISO"] = pd.to_numeric(df_geo["ISO_N3"]) #Делаем столбец ISO одинаковым для обоих датасетов
df_geo = df_geo.drop(labels = "ISO_N3", axis = 1)
for iso in list(set(df_geo["ISO"]) - set(df_sum["ISO"])):
    print(df_geo[df_geo["ISO"] == iso][["ISO", "NAME"]])
```

Из территорий, по которым есть наблюдения, мы потеряли еще и Косово + сомнительные территории, по которым у нас нет наблюдений

```python
df_geo.at[43,"ISO"] = df[df["COUNTRY"] == "France"]["ISO"][0]
df_geo.at[21,"ISO"] = df[df["COUNTRY"] == "Norway"]["ISO"][0]
df_geo.at[174,"ISO"] = df[df["COUNTRY"] == "Kosovo"]["ISO"][0]
```

Соединим снова

Повторим все карты для датасета с протестами.

```python
#Теперь можно делать всё сразу (а лучше вынести в модуль, что я и сделаю)
df2 = pd.read_excel('protests.xlsx')

df2.set_index("EVENT_ID_CNTY", inplace = True)
df2['EVENT_DATE'] = pd.to_datetime(df['EVENT_DATE'])

df_sum2 = df2[["NONPOLITICAL", "ISO"]].groupby("ISO").sum()
df_sum2["DISORDERS"] = df[["NONPOLITICAL", "ISO"]].groupby("ISO").count()
df_sum2["RATIO"] = df_sum2["NONPOLITICAL"] / df_sum2["DISORDERS"]
df_sum2.reset_index(inplace = True)
df_sum2 = pd.DataFrame(df_sum2)
df_sum2["ISO"].astype("int64")

df_geo2 = shape[["ISO_N3", "geometry"]]
df_geo2["ISO"] = pd.to_numeric(df_geo2["ISO_N3"])
df_geo2.drop(labels = "ISO_N3", axis = 1, inplace = True)
df_geo2.at[43,"ISO"] = df2[df2["COUNTRY"] == "France"]["ISO"][0]
df_geo2.at[21,"ISO"] = df2[df2["COUNTRY"] == "Norway"]["ISO"][0]
df_geo2.at[174,"ISO"] = df2[df2["COUNTRY"] == "Kosovo"]["ISO"][0]

df_geo2 = df_geo2.merge(df_sum2, on="ISO")
df_geo2
```

```python
fig, ax = plt.subplots(figsize=(20, 16))
df_geo2.plot(column = "DISORDERS", legend = True, legend_kwds={'shrink': 0.3}, ax=ax)
```

Почти полностью пропал Аравийский полуостров, следовательно, есть совсем экстремальные случаи, когда обычных демонстраций вообще не было.

```python
fig, ax = plt.subplots(figsize=(20, 16))
df_geo2.plot(column = "RATIO", legend = True, legend_kwds={'shrink': 0.3}, ax=ax)
```
