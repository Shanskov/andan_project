```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
df = pd.read_excel('disorders.xlsx')
df.set_index("EVENT_ID_CNTY", inplace = True)
df['EVENT_DATE'] = pd.to_datetime(df['EVENT_DATE'])
df.head()
```

```python
shape = gpd.read_file("mapshape/ne_110m_admin_0_countries.shp")
fig, ax = plt.subplots(figsize=(18, 10))
shape.plot(color="lightgrey", ax=ax)
x = df['LONGITUDE']
y = df['LATITUDE']
plt.scatter(x = "LONGITUDE",y = "LATITUDE", data = df[df['NONPOLITICAL'] == 1])
plt.scatter(x = "LONGITUDE",y = "LATITUDE", data = df[df['NONPOLITICAL'] == 0])
```

```python
df_sum = df[["NONPOLITICAL", "ISO"]].groupby("ISO").sum()
df_sum["DISORDERS"] = df[["NONPOLITICAL", "ISO"]].groupby("ISO").count()
df_sum["RATIO"] = df_sum["NONPOLITICAL"] / df_sum["DISORDERS"]
df_sum.reset_index(inplace = True)
df_sum.columns = ["ISO_N3", "NONPOLITICAL", "DISORDERS", "RATIO"]
df_sum
```
