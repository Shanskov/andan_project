```python
import numpy as np
import pandas as pd

import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns

import SDude as sd #Собственный модуль, подробнее о нем в тетрадках 1 и 1.5

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_excel('datasets/coronavirus_Mar31.xlsx')
df_add = pd.read_csv("datasets/owid-covid-data.csv")
df_fiw = pd.read_excel('datasets/FIW.xlsx', sheet_name="FIW13-23")

useless = ['EVENT_DATE', 'ISO', 'COUNTRY', 'LATITUDE', 'LONGITUDE', 'INTER1',
       'INTER2', 'population', "CL"]

df = sd.dataclear(df, df_add, df_fiw, useless_features=useless, na_limit=0) #Обработка датасетов
del df_add
del df_fiw
```

```python
df1 = df
```

```python
ohe_ready = ["SOURCE_SCALE", 'DISORDER_TYPE', 'EVENT_TYPE', 'SUB_EVENT_TYPE', 'REGION']
```

```python
OneHot = pd.get_dummies(df1[ohe_ready])
df1 = df1.drop(ohe_ready, axis=1)
df1 = df1.join(OneHot)
df1
```

```python
df1 = df1.astype(float)
```

```python
df1.isnull().sum().sort_values(ascending=False)
```

```python
target = "UNORGANIZED"
features = df1.columns.to_list()
features.remove("UNORGANIZED")
```

```python
df_train, df_test = train_test_split(df1, test_size = 0.1, random_state=1, stratify=df1['UNORGANIZED'])
```

```python
df1[target].astype(int).hist()

y_train = df_train[target].values
y_test = df_test[target].values
x_train = df_train[features]
x_test = df_test[features]
```

```python
model = LogisticRegression(max_iter=3000)
model.fit(x_train, y_train)
y_prob = model.predict_proba(x_test)
y_prob = y_prob[:, -1]
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'Accuracy: {accuracy_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Precision: {precision_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Recall: {recall_score(y_test, 1*(y_prob >= 0.5))}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```

```python
y_pred = np.zeros_like(y_test) #Для примера сделаем наивный прогноз
print(f'ACCURACY: {accuracy_score(y_test, y_pred)}')
```

```python
model = LogisticRegression(solver='saga', max_iter=3000)
model.fit(x_train, y_train)
y_prob = model.predict_proba(x_test)
y_prob = y_prob[:, -1]
```

```python
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'Accuracy: {accuracy_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Precision: {precision_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Recall: {recall_score(y_test, 1*(y_prob >= 0.5))}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```

```python
model = LogisticRegression(solver='newton-cholesky', max_iter=1000)
model.fit(x_train, y_train)
y_prob = model.predict_proba(x_test)
y_prob = y_prob[:, -1]
```

```python
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'Accuracy: {accuracy_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Precision: {precision_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Recall: {recall_score(y_test, 1*(y_prob >= 0.5))}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model = LogisticRegression(solver='liblinear', penalty='l1')
model.fit(x_train_scaled, y_train)
y_prob = model.predict_proba(x_test_scaled)
y_prob = y_prob[:, -1]
```

```python
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'Accuracy: {accuracy_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Precision: {precision_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Recall: {recall_score(y_test, 1*(y_prob >= 0.5))}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```

```python
model = LogisticRegression(max_iter=1000)

from sklearn.model_selection import GridSearchCV 
param_grid = {'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']} #Доступные солверы
grid_search = GridSearchCV(model, param_grid, cv=4, scoring="roc_auc")

grid_search.fit(x_train_scaled, y_train)

grid_search.best_estimator_
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model = model.fit(x_train, y_train)
y_prob = model.predict_proba(x_test)[:,-1]
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```

```python
from catboost import CatBoostClassifier

model = CatBoostClassifier(iterations=1000, random_seed=123)
model.fit(x_train, y_train)
y_prob = model.predict_proba(x_test)[:,-1]
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```

```python
from sklearn.svm import LinearSVC
svm = LinearSVC(dual=False)
svm.fit(x_train, y_train)
svm_pred = svm.predict(x_test)
roc_auc_score(y_test, svm_pred)
```
