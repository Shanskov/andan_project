# Проверка гипотез
У нас есть несколько столбцов с информацией о самой новости, которые могут помочь в определении таргета. Если с типом события ("EVENT_TYPE" и др.) всё и так понятно - это важная информация, то с источником новости всё неочевидно, и стоит проветить такую гипотезу:
$$
\begin{aligned}
& H_0: \text{"SOURCE_SCALE" и "UNORGANIZED" независимы} \\
& H_1: \text{"SOURCE_SCALE" и "UNORGANIZED" зависимы}
\end{aligned}
$$
<br>
Проверка корреляции масштаба источника и организованности события, выданная ChatGPT.

Запрос: I have a pandas dataframe where "SOURCE_SCALE" column contains categorial feature, and another column named "UNORGANIZED", with boolean values. I want to know if there is a correlation between categorial feature and bool value.

```python
# Критерий согласия Пирсона, ничего интересного.

# create contingency table

contingency_table = pd.crosstab(df['SOURCE_SCALE'], df['UNORGANIZED'])

# perform chi-square test of independence

chi2, p_value, _, _ = sts.chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("p-value:", p_value)
```

Получаем, что эти два столбца точно зависят друг от друга. Посчитаем доли неорганизованных событий для каждого типа источника и построим картинку.

```python
df_hypo = df[["SOURCE_SCALE","UNORGANIZED"]].groupby(by = "SOURCE_SCALE").mean()
df_hypo.plot(kind = "bar");
```

Странные значения объясняются тем, что из некоторых типов источников поступило очень мало новостей. Вот из Regional источника есть только одна новость во всём очищенном датасете.

```python
df[df["SOURCE_SCALE"] == "Regional"]
```

```python
df["SOURCE_SCALE"].value_counts()
```

<b>Итог: </b> гипотеза подтверждается, "SOURCE_SCALE" - важный признак, который нельзя выкидывать.

### А не временной ли у нас ряд?
Если вдруг для это так, то это будет ОЧЕНЬ ПЕЧАЛЬНО. Временные ряды - слишком серьезное препятствие, чтобы хорошо работать с ними одному и без достаточного опыта.

```python
dtime = df[["EVENT_DATE", "UNORGANIZED"]].groupby(by = ["EVENT_DATE"]).count()
plt.figure(figsize = (16, 12))
plt.plot(dtime)
```

```python
#Подсмотренная на хабре реализация расширенного теста Дики - Фуллера, источник: https://habr.com/ru/articles/207160/
import statsmodels.api as sm
test = sm.tsa.adfuller(dtime)
print('adf: ', test[0])
print('p-value: ', test[1])
print('Critical values: ', test[4])
if test[0]> test[4]['5%']: 
    print('есть единичные корни, ряд не стационарен')
else:
    print('единичных корней нет, ряд стационарен')
```

Так, количество новостей по дням - стационарный ряд, но что насчет доли неорганизованных событий?

```python
unorg_ratio = df[["EVENT_DATE", "UNORGANIZED"]].groupby(by = ["EVENT_DATE"]).mean()
unorg_ratio
plt.figure(figsize = (16, 12))
plt.plot(unorg_ratio);
test = sm.tsa.adfuller(unorg_ratio)
print('adf: ', test[0])
print('p-value: ', test[1])
print('Critical values: ', test[4])
if test[0]> test[4]['5%']: 
    print('есть единичные корни, ряд не стационарен')
else:
    print('единичных корней нет, ряд стационарен')
```

<b>Итог: </b> гипотеза отвергается, ряд стационарный. <br>
А если посмотреть на страны отдельно?

```python
n_obs = 0
for country in df["COUNTRY"].unique():
    try:
        test = sm.tsa.adfuller(df[df["COUNTRY"] == country][["EVENT_DATE", "UNORGANIZED"]].groupby(by = ["EVENT_DATE"]).mean())
        if test[0]> test[4]['5%']:
            n_obs += df[df["COUNTRY"] == country]["UNORGANIZED"].count()
            print(country, df[df["COUNTRY"] == country]["UNORGANIZED"].count())
    except ValueError:
        next
print(n_obs)   
```

К сожалению, не у всех стран всё стационарно, НО у них в сумме не очень много наблюдений, поэтому я бы даже мог их выкинуть (но не буду). В дальнейшем это не должно помешать.
