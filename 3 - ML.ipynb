```python
import numpy as np
import pandas as pd

import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns

import SDude as sd #Собственный модуль, подробнее о нем в тетрадках 1 и 1.5

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
```

# Создание датасета и обработка

```python
df = pd.read_excel('datasets/coronavirus_Mar31.xlsx')
df_add = pd.read_csv("datasets/owid-covid-data.csv")
df_fiw = pd.read_excel('datasets/FIW.xlsx', sheet_name="FIW13-23")

useless = ['EVENT_DATE', 'ISO', 'LATITUDE', 'LONGITUDE', 'INTER1',
       'INTER2', 'population', "CL", "PR", "Total"]

df = sd.dataclear(df, df_add, df_fiw, useless_features=useless, na_limit=5) #Обработка датасетов
del df_add
del df_fiw
df
```

```python
df.columns
```

```python
ohe_ready = ["SOURCE_SCALE", 'DISORDER_TYPE', 'EVENT_TYPE', 'SUB_EVENT_TYPE', 'REGION']
```

```python
OneHot = pd.get_dummies(df[ohe_ready])
df = df.drop(ohe_ready, axis=1)
df = df.join(OneHot)
df
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df["death/case ratio"] = df["total_deaths"] / df["total_cases"]
df = df.drop(["total_deaths", "total_cases", 'total_cases_per_million', 'total_deaths_per_million'], axis=1)
```

```python
for key in df.isnull().sum().sort_values(ascending=False).keys()[:18]:
    df[key] = df.groupby('COUNTRY')[key].transform(lambda x: x.fillna(x.mean()))
```

```python
df1 = df.dropna()
df1 = df1.drop(["COUNTRY"], axis=1)
```

```python
df1
```

```python
df1 = df1.astype(float)
```

# Машинное обучение

```python
target = "UNORGANIZED"
features = df1.columns.to_list()
features.remove("UNORGANIZED")
```

```python
df_train, df_test = train_test_split(df1, test_size = 0.1, random_state=1, stratify=df1['UNORGANIZED'])
```

```python
y_train = df_train[target].values
y_test = df_test[target].values
x_train = df_train[features]
x_test = df_test[features]
```

Начинаем с нашего любимого логрега.

```python
model = LogisticRegression(solver='newton-cholesky', max_iter=1000)
model.fit(x_train, y_train)
y_prob = model.predict_proba(x_test)
y_prob = y_prob[:, -1]
```

```python
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'Accuracy: {accuracy_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Precision: {precision_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Recall: {recall_score(y_test, 1*(y_prob >= 0.5))}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```

Очень неплохо, но лучше не будет - я проверял, поэтому попробуем нормализовать данные.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model = LogisticRegression(solver='newton-cholesky', max_iter=1000)
model.fit(x_train_scaled, y_train)
y_prob = model.predict_proba(x_test_scaled)
y_prob = y_prob[:, -1]
```

```python
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'Accuracy: {accuracy_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Precision: {precision_score(y_test, 1*(y_prob >= 0.5))}')
print(f'Recall: {recall_score(y_test, 1*(y_prob >= 0.5))}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```

Никаких изменений, это печально. Перейдем к другим методам.

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model = model.fit(x_train_scaled, y_train)
y_prob = model.predict_proba(x_test_scaled)[:,-1]
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```

Решающие пни как всегда хороши! Можно было бы остановиться, но есть же еще кэтбуст!

```python
from catboost import CatBoostClassifier

model = CatBoostClassifier(iterations=1000, random_seed=123)
model.fit(x_train_scaled, y_train)
y_prob = model.predict_proba(x_test_scaled)[:,-1]
print(f'ROC-AUC: {roc_auc_score(y_test, y_prob)}')
print(f'F1: {f1_score(y_test, 1*(y_prob >= 0.5))}')
```
